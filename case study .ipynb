```python
f = open("thuvien.csv","r",encoding = "utf8")
header = f.readline() 
data = f.readlines()

D_data = {}

header = header.strip().split(",")



for i in header:
    D_data[i] = []

for i in data:
    i = i.strip().split(",")
    index = 0
    for key in D_data:
            D_data[key].append(i[index])
            index += 1

            
L_thongtin=[]
for a in data:
    L_thongtin.append(a)

    
    
from tkinter import *
from tkinter import messagebox
sc = Tk()
sc.title('quản lý mượn trả sách')

def change():
    fr2.pack_forget()
    fr1.pack()
def change1():
    fr1.pack_forget()
    fr2.pack()

def luu():
    ten = et1_var.get()
    lop = et2_var.get()
    sach = et3_var.get()
    tacgia = et4_var.get()
    ma = et5_var.get()
    muon = et6_var.get()
    f = open("thuvien.csv",'a',encoding = 'utf8') 
    f.write(ten +','+lop+','+sach+','+ tacgia+','+ma+','+ muon+','+ "\n")
    f.close()
    D_data['họ tên'].append(et1_var.get())
    D_data['mã sách'].append(et5_var.get())
    et1_var.set('')
    et2_var.set('')
    et3_var.set('')
    et4_var.set('')
    et5_var.set('')
    et6_var.set('')
    return messagebox.showinfo("đã lưu thông tin")

def timkiem():
    if et12_var.get() not in D_data['họ tên']:
        lb7_fr2.config(text='không tồn tại tên người mượn')
        return
    vt = D_data['họ tên'].index(et12_var.get())
    if D_data['mã sách'][vt] != et22_var.get():
        lb7_fr2.config(text = "không tồn tại")
        return
    lb7_fr2.config(text=L_thongtin[vt])

def ghifile():
    vt = D_data['họ tên'].index(et12_var.get())
    D_data['ngày trả'][vt] = et26_var.get()
    f = open("thuvien.csv","w",encoding = "utf8")
    chuoi_header = ",".join(header)
    f.write(chuoi_header + "\n")
    for i in range(len(D_data['ngày trả'])):
        chuoi = ""
        for key in D_data:
            chuoi += D_data[key][i] + ","
        f.write(chuoi[:-1] + "\n")
    f.close()
    return messagebox.showinfo("đã cập nhật thông tin")


### thiết kế giao diện



fr0=LabelFrame()

lb1 = Label(fr0,text= 'quản lý mượn trả sách', font=('Arial',15),width = 30)
bt1= Button(fr0, text ='mượn sách',font =('Arial',15),width=30,command =change)
bt2= Button(fr0, text ='Trả sách',font =('Arial',15),width=30,command =change1)

lb1.grid(row=1,column=1,columnspan=2)
bt1.grid(row=2,column=1)
bt2.grid(row=2,column=2)

fr0.pack()




fr1 = LabelFrame()
lb1 = Label(fr1,text='MƯỢN SÁCH', font=('Arial',20),width = 20)
hoten= Label(fr1,text='Họ và Tên',font=("Arial",15),width = 20)
et1_var= StringVar()
et1= Entry(fr1,textvariable = et1_var,font=("Arial",15),width = 50)

lop= Label(fr1, text ='Lớp', font=('Arial',15),width = 20)
et2_var= StringVar()
et2= Entry (fr1, textvariable= et2_var, font=('Arial',15),width = 50)

sach = Label(fr1,text= 'Tên sách', font=('Arial',15),width = 20)
et3_var= StringVar()
et3= Entry (fr1, textvariable= et3_var, font=('Arial',15),width = 50)

tacgia= Label(fr1, text ='Tác giả', font=('Arial',15),width = 20)
et4_var= StringVar()
et4= Entry (fr1, textvariable= et4_var, font=('Arial',15),width = 50)

ma= Label(fr1, text ='Mã sách', font=('Arial',15),width = 20)
et5_var= StringVar()
et5= Entry (fr1, textvariable= et5_var, font=('Arial',15),width = 50)

lb_ngaymuon= Label(fr1, text ='Ngày mượn sách', font=('Arial',15),width = 20)
et6_var= StringVar()
et6= Entry (fr1, textvariable= et6_var, font=('Arial',15),width = 50)

bt1= Button(fr1,text ='LƯU', font=('Arial',15),width=10,command=luu)


lb1.grid(row=1,column=1,columnspan=2)
hoten.grid(row=2,column=1)
et1.grid(row=2,column=2)
lop.grid(row=3,column=1)
et2.grid(row=3,column=2)
sach.grid(row=4,column=1)
et3.grid(row=4,column=2)
tacgia.grid(row=5,column=1)
et4.grid(row=5,column=2)
ma.grid(row=6,column=1)
et5.grid(row=6,column=2)
lb_ngaymuon.grid(row=7,column=1)
et6.grid(row=7,column=2)
bt1.grid(row=8,column=1,columnspan=2)



fr2 = LabelFrame()
lb1_fr2 = Label(fr2,text='TRẢ SÁCH', font=('Arial',20),width = 20)
hoten_fr2= Label(fr2,text='Họ và Tên',font=("Arial",15),width = 20)
et12_var= StringVar()
et12= Entry(fr2,textvariable = et12_var,font=("Arial",15),width = 20)
ma_fr2= Label(fr2, text ='Mã sách', font=('Arial',15),width = 20)
et22_var= StringVar()
et22= Entry (fr2, textvariable= et22_var, font=('Arial',15),width = 20)
lb7_fr2 = Label(fr2,text='', font=('Arial',15),width = 100)
bt2= Button(fr2, text ='Tìm kiếm',font =('Arial',15),width=20,command= timkiem)
ngaytra_fr2= Label(fr2, text ='ngày trả sách', font=('Arial',15),width = 20)
et26_var= StringVar()
et26= Entry (fr2, textvariable= et26_var, font=('Arial',15),width = 20)
bt22= Button(fr2,text ='xác nhận', font=('Arial',15),width=10,command= ghifile) 

lb1_fr2.grid(row=1,column=1,columnspan=2)
hoten_fr2.grid(row=2,column=1)
et12.grid(row=2,column=2)
ma_fr2.grid(row=3,column=1)
et22.grid(row=3,column=2)
lb7_fr2.grid(row=4,column=1,columnspan=2)
bt2.grid(row=5, column=1,columnspan=2)
ngaytra_fr2.grid(row=6,column=1)
et26.grid(row=6,column=2)
bt22.grid(row=7,column=1,columnspan=2)



    

sc.mainloop()
```
